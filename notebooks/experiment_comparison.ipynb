# Philosophical Engine - Experiment Comparison

This notebook connects to the local MLflow server, fetches all runs across the classification and regression experiments, and displays a summary table of model metrics.

```python
import mlflow
import pandas as pd
import os

# Set tracking URI
mlflow.set_tracking_uri(os.getenv("MLFLOW_TRACKING_URI", "http://localhost:5000"))

print(f"Connected to MLflow at {mlflow.get_tracking_uri()}")
```

### Classification Models

```python
class_exp = mlflow.get_experiment_by_name("philosophical-engine-classification")
if class_exp:
    df_class = mlflow.search_runs([class_exp.experiment_id])
    if not df_class.empty:
        display_cols = ['tags.mlflow.runName', 'metrics.accuracy', 'metrics.f1_macro', 'params.model']
        # Filter columns that exist
        cols = [c for c in display_cols if c in df_class.columns]
        display(df_class[cols].sort_values(by='metrics.f1_macro', ascending=False))
else:
    print("No classification experiment found.")
```

### Regression Models

```python
reg_exp = mlflow.get_experiment_by_name("philosophical-engine-regression")
if reg_exp:
    df_reg = mlflow.search_runs([reg_exp.experiment_id])
    if not df_reg.empty:
        display_cols = ['tags.mlflow.runName', 'metrics.r2', 'metrics.rmse', 'metrics.mae', 'params.model']
        cols = [c for c in display_cols if c in df_reg.columns]
        display(df_reg[cols].sort_values(by='metrics.r2', ascending=False))
else:
    print("No regression experiment found.")
```
